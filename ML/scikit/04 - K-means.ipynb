```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets
```

```python
iris = datasets.load_iris()

x = pd.DataFrame(iris.data)
x.columns = ['Sepal_length','Sepal_width','Petal_length','Petal_width']
y = pd.DataFrame(iris.target)
y.columns = ['Targets']

print(x[:10])
```

```python
# http://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html
model = KMeans(n_clusters=3)
model.fit(x)
```

```python
colormap = np.array(['red','green','blue'])
colormap2 = np.array(['blue','red','green'])

plt.figure(figsize=(10,4))

plt.subplot(1, 2, 1)
plt.scatter(x.Petal_length, x.Petal_width, c=colormap[y.Targets.values], s=40)
plt.title('Labels (truth)')

plt.subplot(1, 2, 2)
plt.scatter(x.Petal_length, x.Petal_width, c=colormap2[model.labels_], s=40)
plt.title('Clusters (prediction)')

plt.tight_layout()
plt.show()
```
